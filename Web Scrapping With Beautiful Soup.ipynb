```python
# this module helps in web scrapping.
from bs4 import BeautifulSoup 

# this module helps us to download a web page
import requests 
```

```python
# Provide the website with information to be scrapped
website = "https://www.allaboutbirds.org/guide/browse/taxonomy"
```

```python
# To get the html components from the website, a GET request is sent
components = requests.get(website).text

# To parse the website component documents from the GET request above
documents = BeautifulSoup(components,"html.parser")
```

```python
# Check the documents in a nested HTML structure like you would see in the once you click inspect on the website
print(documents.prettify())
```

```python
# Check for the file title
print(documents.title.text)
```

```python
for link in documents.find_all(["a","h2"]):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
    print("h2:{}".format(link.get("h2")))
```

```python
# Finds and extracts all div with class housing information of species
species = documents.findAll('div', attrs={'class':'species-info'})

# Looping through the div, selecting all anchor tag text for species family and type , selecting not just the links of href
species_type = []
for div in species:
    species_type.append(div.find('a').contents[0])    
    
#print(species_type)   


species_info = []
for div in species:
    species_info.append(div.find('em').contents[0])    
    
#print(species_family) 


# Add Index column
count = 0
index_list = []
for i in species_type:
    count = count+1
    index_list.append(count)

# Substract 1 from every value in the index_list
index_list[:] = [value - 1 for value in index_list]

# Create a dataframe for species type by species info
import pandas as pd
dictionary = {"Bird Type": species_type, "Bird Info": species_info, "Index": index_list}
bird_table = pd.DataFrame(dictionary)

bird_table.reset_index()

bird_table
```

```python
# Finds and extracts all div with class housing information of species order: 21 Different birds order
species_order = documents.findAll('div', attrs={'class':'order-name'})

species_order_name = []
for div in species_order:
    species_order_name.append(div.find('h2').contents[0])  
    


dictionary = {"Bird Order": species_order_name}
order_table = pd.DataFrame(dictionary)

order_table
```

```python
# Finds and extracts all div with class housing information of species family
species_family = documents.findAll('div', attrs={'class':'family-name sticky'})


species_family_name = []
for div in species_family:
    species_family_name.append(div.find('a').contents[0])  
```
